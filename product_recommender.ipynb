## Create user-product matrix

```python
user_product_piv = item_details.pivot_table(index = 'Transaction ID', columns = 'Product', values = 'Quantity')
# users_artists_piv = users_artists.pivot(index = 'userID', columns = 'name', values = 'plays').fillna(0)
# users_artists_piv.head()
```

```python
user_product_piv = user_product_piv.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
```

```python
distance_matrix = cosine_similarity(user_product_piv.T)
distance_matrix.shape
```

```python
distDF = pd.DataFrame(distance_matrix, columns=(user_product_piv.columns), index=(user_product_piv.columns))
distDF.head()
```

```python
distDF_melt = pd.DataFrame(distance_matrix, columns=(user_product_piv.columns), index=(user_product_piv.columns))
distDF_melt = distDF.reset_index()
distDF_melt.head()
```

```python
top_5_recommended = pd.DataFrame(columns=list(range(0,6)))
```

```python
distDF_binary_idx = distDF.reset_index()
```

```python
distDF_binary_idx
```

```python
distDF_binary_idx.columns
```

```python
for idx, row in distDF_binary_idx.iterrows():
    products = []
    products.append(row['Product'])
    top_5 = row[1:].sort_values(ascending=False)[1:6]
    for product, dist in top_5.iteritems():
        products.append(product)
    top_5_recommended.loc[idx] = products
    
```
