```python
items = list(transactions.Product)
combos = []
for transaction in items:
    if len(transaction) == 1:
        single_transaction = (transaction[0], transaction[0])
        combos.append(single_transaction)
    else:
        combos.append([(transaction[i],transaction[j]) for i in range(len(transaction)) for j in range(i+1, len(transaction))])
    
    
```

```python
items = list(transactions.Product)
multiproduct_combos = []
for transaction in items:
    if len(transaction) > 1:
#         for combinations in transaction:
#             multiproduct_combos.append([(combinations[i],combinations[j]) for i in range(len(combinations)) for j in range(i+1, len(combinations))])
        multiproduct_combos.append([(transaction[i],transaction[j]) for i in range(len(transaction)) for j in range(i+1, len(transaction))])
    
    
```

```python
pd.DataFrame([j for i in multiproduct_combos for j in i], columns = ['item 1', 'item 2']).groupby(['item 1','item 2']).size().reset_index().rename(columns={0:'count'})
```

```python
multiproduct_orders = pd.DataFrame([j for i in multiproduct_combos for j in i], columns = ['item 1', 'item 2']).groupby(['item 1','item 2']).size().reset_index().rename(columns={0:'count'})
multiproduct_orders.to_csv('multiproduct_orders.csv')
```

```python
all_combos = []
for order in combos:
    if type(order) == list:
        for pair in order:
            all_combos.append(pair)
    else:
        all_combos.append(order)
```

```python
all_combos = pd.DataFrame(all_combos, columns = ['item 1', 'item 2'])
```

```python
all_combos.to_csv('combos_for_fusiontable.csv')
```
